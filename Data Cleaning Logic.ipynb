```
import pandas as pd
import numpy as np
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
import seaborn as sns
import matplotlib.pyplot as plt


print("LOADING DATA...")
day_df = pd.read_csv('/content/day.csv')
hour_df = pd.read_csv('/content/hour.csv')


def analyze_and_clean_duplicates(df, name):
    print(f"\n{'='*20} ANALYSIS FOR: {name} {'='*20}")


    print(f"\n--- First 5 Rows ({name}) ---")
    display(df.head())


    print(f"\n--- Info ({name}) ---")
    df.info()


    print(f"\n--- Missing Values ({name}) ---")
    print(df.isnull().sum())


    print(f"\n--- Statistics ({name}) ---")
    display(df.describe())

    # Drop Duplicates
    initial_count = len(df)
    df = df.drop_duplicates()
    new_count = len(df)
    if initial_count != new_count:
        print(f"\n Duplicates Removed: {initial_count - new_count} rows dropped.")
    else:
        print("\n No duplicates found.")


    num_features = df.select_dtypes(include=[np.number]).columns.tolist()
    cat_features = df.select_dtypes(exclude=[np.number]).columns.tolist()

    print(f"\nNumerical Features: {num_features}")
    print(f"Categorical Features: {cat_features}")

    return df
#call both function
day_df = analyze_and_clean_duplicates(day_df, "Day Dataset")
hour_df = analyze_and_clean_duplicates(hour_df, "Hour Dataset")







print(f"\n{'='*20} APPLYING TRANSFORMATION {'='*20}")

new_names = {
    'yr': 'year',
    'mnth': 'month',
    'hum': 'humidity',
    'cnt': 'total_count',
    'weathersit': 'weather_condition',
    'hr': 'hour'
}
season_names = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
weather_names = {1: 'Clear', 2: 'Mist', 3: 'Light Snow', 4: 'Heavy Rain'}

def clean_transform(df):
    df = df.copy()
    # Rename
    df.rename(columns=new_names, inplace=True)

    # Date conversion
    if 'dteday' in df.columns:
        df['dteday'] = pd.to_datetime(df['dteday'])

    # Value Mapping
    if 'season' in df.columns:
        df['season'] = df['season'].map(season_names)
    if 'weather_condition' in df.columns:
        df['weather_condition'] = df['weather_condition'].map(weather_names)

    return df

day_cleaned = clean_transform(day_df)
hour_cleaned = clean_transform(hour_df)

print("\n Transformation Complete!")
print("Cleaned Day Data Head:")
display(day_cleaned.head())




day_cleaned.to_csv('day_cleaned.csv', index=False)
hour_cleaned.to_csv('hour_cleaned.csv', index=False)

print("\n Downloading files...")
files.download('day_cleaned.csv')
files.download('hour_cleaned.csv')
```
