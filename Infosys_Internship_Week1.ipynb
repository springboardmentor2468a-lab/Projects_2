**Merging Both Datasets**

```python
# Convert dteday to datetime type
dfh['dteday'] = pd.to_datetime(dfh['dteday'])
dfd['dteday'] = pd.to_datetime(dfd['dteday'])
```

```python
# Standardize column types
dfh['season'] = dfh['season'].astype(int)
dfd['season'] = dfd['season'].astype(int)

merge_columns = ['dteday', 'temp', 'atemp', 'hum', 'windspeed']

df_preprocessed = dfh.merge(
    dfd[merge_columns],
    on='dteday',
    how='left',
    suffixes=('_hour', '_day')
)

print("Merged dataset successfully!")
df_preprocessed.head()
```

```python
df_preprocessed.info()
```

**Dropped repeated or unnecessary columns**

```python
columns_to_drop = [
    'dteday',
    'hr',
    'casual',
    'registered'
]

df_preprocessed = df_preprocessed.drop(columns=columns_to_drop, errors='ignore')
```

```python
df_preprocessed['date'] = pd.to_datetime(df_preprocessed['date'])

print("Cleaned columns:")
print(df_preprocessed.info())
```

**Rearranged dataset and imported to .csv file**

```python
new_order = [
    'datetime', 'date', 'hour',      # time features

    # categorical features
    'season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit',

    # numerical features
    'temp_hour', 'atemp_hour', 'hum_hour', 'windspeed_hour',

    # numerical features
    'temp_day', 'atemp_day', 'hum_day', 'windspeed_day',

    # target variable
    'cnt'
]

# Rearrange dataset
df_ready = df_preprocessed[new_order]
df_ready.to_csv("Cleaned_Bikeshare_Data.csv", index=False)

print("Dataset rearranged and saved as cleaned_bikeshare_data.csv")
```

```python
print("Final Dataset:")
df_ready.head()
```

**Visualization Of Bike Rental Count Hourly and Monthly**

```python
plt.figure(figsize=(12,6))
sns.lineplot(x="hour", y="cnt", data=df_ready, estimator="mean")
plt.title("Average Bike Rentals by Hour of the Day")
plt.xlabel("Hour")
plt.ylabel("Average Rental Count")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x="mnth", y="cnt", data=df_ready, estimator="mean", marker="o")
plt.title("Average Monthly Bike Rentals")
plt.xlabel("Month")
plt.ylabel("Average Rental Count")
plt.grid(True)
plt.show()
```
