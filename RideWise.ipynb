# Task 1 = Load Dataset

```python
import pandas as pd 

df_day = pd.read_csv("day.csv")
df_hour = pd.read_csv("hour.csv")
```

# Task -2 EDA Analysis

```python
print("DAY Dataset Shape:", df_day.shape)
print("HOUR Dataset Shape:", df_hour.shape)
```

```python
print("\nDAY Columns:\n", df_day.columns)
print("\nHOUR Columns:\n", df_hour.columns)
```

```python
print("\nDAY Info:")
print(df_day.info())

print("\nHOUR Info:")
print(df_hour.info())
```

```python
print("\nDAY Null Values:")
print(df_day.isnull().sum())

print("\nHOUR Null Values:")
print(df_hour.isnull().sum())
```

```python
print("\nDAY Null Values:")
print(df_day.isnull().sum().sum())

print("\nHOUR Null Values:")
print(df_hour.isnull().sum().sum())
```

```python
print("\nDAY Duplicate Rows:", df_day.duplicated().sum())
print("HOUR Duplicate Rows:", df_hour.duplicated().sum())
```

```python
print("\nDAY Statistics:\n", df_day.describe())
print("\nHOUR Statistics:\n", df_hour.describe())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline


plt.figure(figsize=(10,5))
df_day.isnull().sum().plot(kind='bar', title='Missing Values in Day Dataset')
# plt.show()

plt.figure(figsize=(10,5))
df_hour.isnull().sum().plot(kind='bar', title='Missing Values in Hour Dataset')
# plt.show()
```

```python
plt.figure(figsize=(10,5))
df_day.isnull().sum().plot(kind='bar', title='Missing Values in Day Dataset')


plt.figure(figsize=(10,5))
df_hour.isnull().sum().plot(kind='bar', title='Missing Values in Hour Dataset')

```

```python


 num_cols_day = ['instant', 'dteday', 'season', 'yr', 'mnth', 'holiday', 'weekday',
       'workingday', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed',
       'casual', 'registered', 'cnt']
     


num_cols_hour=['instant', 'dteday', 'season', 'yr', 'mnth', 'hr', 'holiday', 'weekday',
       'workingday', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed',
       'casual', 'registered', 'cnt']
     
for col in num_cols_day:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df_day[col])
    plt.title(f"Day Dataset - Boxplot of {col}")
    plt.show()

for col in num_cols_hour:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df_hour[col])
    plt.title(f"Hour Dataset - Boxplot of {col}")
    plt.show()
```
