```python
import pandas as pd
import numpy as np
day_df = pd.read_csv("/content/day.csv")
hour_df = pd.read_csv("/content/hour.csv")
day_df.columns = day_df.columns.str.lower().str.strip().str.replace(" ", "_")
hour_df.columns = hour_df.columns.str.lower().str.strip().str.replace(" ", "_")
numeric_day = day_df.select_dtypes(include=[np.number]).columns
numeric_hour = hour_df.select_dtypes(include=[np.number]).columns

day_df[numeric_day] = day_df[numeric_day].fillna(day_df[numeric_day].median())
hour_df[numeric_hour] = hour_df[numeric_hour].fillna(hour_df[numeric_hour].median())

categorical_day = day_df.select_dtypes(include=['object']).columns
categorical_hour = hour_df.select_dtypes(include=['object']).columns

for col in categorical_day:
    day_df[col] = day_df[col].fillna(day_df[col].mode()[0])

for col in categorical_hour:
    hour_df[col] = hour_df[col].fillna(hour_df[col].mode()[0])

day_df.drop_duplicates(inplace=True)
hour_df.drop_duplicates(inplace=True)

if "temp" in day_df.columns:
    day_df = day_df[(day_df["temp"] >= 0) & (day_df["temp"] <= 1)]

if "hum" in day_df.columns:
    day_df = day_df[(day_df["hum"] >= 0) & (day_df["hum"] <= 1)]

if "windspeed" in day_df.columns:
    day_df = day_df[(day_df["windspeed"] >= 0)]

if "temp" in hour_df.columns:
    hour_df = hour_df[(hour_df["temp"] >= 0) & (hour_df["temp"] <= 1)]

if "hum" in hour_df.columns:
    hour_df = hour_df[(hour_df["hum"] >= 0) & (hour_df["hum"] <= 1)]

if "windspeed" in hour_df.columns:
    hour_df = hour_df[(hour_df["windspeed"] >= 0)]
print("Day dataset:", day_df.shape)
print("Hour dataset:", hour_df.shape)
```

```python
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])

merged_df = hour_df.merge(
    day_df,
    on=['dteday', 'season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit'],
    suffixes=('_hour', '_day')
)

print("Merged dataset shape:", merged_df.shape)
merged_df.head()
```

```python

merged_df['dteday'] = pd.to_datetime(merged_df['dteday'])
merged_df['year'] = merged_df['dteday'].dt.year
merged_df['month'] = merged_df['dteday'].dt.month
merged_df['day'] = merged_df['dteday'].dt.day
merged_df['weekday'] = merged_df['dteday'].dt.weekday
merged_df['hour'] = merged_df['hr']
merged_df['is_weekend'] = merged_df['weekday'].isin([5, 6]).astype(int)
merged_df['weeknum'] = merged_df['dteday'].dt.isocalendar().week.astype(int)
merged_df['weekday_name'] = merged_df['dteday'].dt.day_name()
merged_df['temp_level'] = pd.cut(
    merged_df['temp_hour'],
    bins=[0, 0.33, 0.66, 1],
    labels=['Low', 'Medium', 'High']
)
merged_df['humid_flag'] = (merged_df['hum_hour'] > merged_df['hum_hour'].mean()).astype(int)
merged_df['windy_flag'] = (merged_df['windspeed_hour'] > merged_df['windspeed_hour'].mean()).astype(int)
merged_df['weather_severity'] = (
    merged_df['temp_hour'] * 0.4 +
    merged_df['hum_hour'] * 0.3 +
    merged_df['windspeed_hour'] * 0.3
)
merged_df['is_holiday'] = merged_df['holiday']
merged_df['is_event_day'] = ((merged_df['holiday'] == 1) | (merged_df['workingday'] == 0)).astype(int)
merged_df['major_event_flag'] = (
    (merged_df['weathersit'] >= 3) |
    (merged_df['holiday'] == 1)
).astype(int)
merged_df['cnt_lag_1'] = merged_df['cnt_day'].shift(1)

merged_df['cnt_7day_avg'] = merged_df['cnt_day'].rolling(window=7).mean()

merged_df = merged_df.dropna().reset_index(drop=True)
merged_df.head(), merged_df.shape
```
