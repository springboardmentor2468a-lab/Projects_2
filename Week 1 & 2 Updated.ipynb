```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_d = data_d.drop(columns=['instant'])
data_h = data_h.drop(columns= ['instant'])
```

```python
print(data_d.columns, "day")
print(data_h.columns, "hour")
```

```python
# Distribution of total bike reentals
plt.figure(figsize=(7,5))
sns.histplot(data_d['cnt'], kde=True)
plt.title("Distribution of Total Bike Rentals")
plt.show()
```

```python
# Monthly rental count
monthly_counts = data_d.groupby('mnth')['cnt'].sum().reset_index()

monthly_counts.style.background_gradient(
    subset=['cnt'], cmap='YlGnBu'
)
```

```python
# Bike rentals over the time 
plt.figure(figsize=(12,5))
plt.plot(data_d['dteday'], data_d['cnt'])
plt.title("Bike Rentals Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.show()
```

```python
# Average rentals by the season
plt.figure(figsize=(7,5))
sns.barplot(x='season', y='cnt', data=data_d)
plt.title("Average Rentals by Season")
plt.show()
```

```python
# Rentals on weekdays and working days
sns.barplot(x='workingday', y='cnt', data=data_d)
plt.title("Effect of Working Day on Rentals")
plt.show()
```

```python
# Scatter plot for Temp vs Rentals
plt.figure(figsize=(7,5))
sns.scatterplot(x='temp', y='cnt', data=data_d)
plt.title("Temperature vs Rentals")
plt.show()
```
