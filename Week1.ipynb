```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
hour.hist(figsize=(12,8))
plt.show()
```

```python
day.hist(figsize=(12,8))
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(hour.select_dtypes(include=['int64','float64']).corr(), cmap="coolwarm")
plt.title("Hour Dataset Correlation")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(day.select_dtypes(include=['int64','float64']).corr(), cmap="coolwarm")
plt.title("Day Dataset Correlation")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="season", y="cnt", data=day)
plt.title("Season vs Count")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(day['dteday'], day['cnt'])
plt.title("Daily Bike Rentals Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="temp", y="cnt", data=hour)
plt.title("Temperature vs Count")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(x="hr", y="cnt", data=hour)
plt.title("Hourly Trend of Bike Rentals")
plt.show()
```
