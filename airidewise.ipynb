```python
# Check for impossible or inconsistent values

print("Negative temperature values:", (data['temp'] < 0).sum())
print("Negative atemp values:", (data['atemp'] < 0).sum())
print("Negative humidity values:", (data['hum'] < 0).sum())
print("Humidity > 1 values:", (data['hum'] > 1).sum())
print("Negative windspeed values:", (data['windspeed'] < 0).sum())
print("Hours outside 0–23:", (~data['hr'].between(0, 23)).sum())
print("Months outside 1–12:", (~data['mnth'].between(1, 12)).sum())
```

```python
print("Invalid date entries:", pd.to_datetime(data['dteday'], errors='coerce').isna().sum())
```
