```python
import pandas as pd

day = pd.read_csv("day.csv")
hour = pd.read_csv("hour.csv")

print("DAY CSV:")
display(day.head())

print("HOUR CSV:")
display(hour.head())
```

```python
print("Missing in DAY:")
display(day.isna().sum())

print("Missing in HOUR:")
display(hour.isna().sum())
```

```python
day["_source"] = "day"
hour["_source"] = "hour"

merged = pd.concat([day, hour], ignore_index=True)
display(merged.head())
```
