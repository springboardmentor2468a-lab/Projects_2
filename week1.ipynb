```python
# Check for impossible or inconsistent values

print("Negative temperature values:", (hour['temp'] < 0).sum())
print("Negative atemp values:", (hour['atemp'] < 0).sum())
print("Negative humidity values:", (hour['hum'] < 0).sum())
print("Humidity > 1 values:", (hour['hum'] > 1).sum())
print("Negative windspeed values:", (hour['windspeed'] < 0).sum())
print("Hours outside 0–23:", (~hour['hr'].between(0, 23)).sum())
print("Months outside 1–12:", (~hour['mnth'].between(1, 12)).sum())
```

```python
print("Invalid date entries:", pd.to_datetime(hour['dteday'], errors='coerce').isna().sum())
```
